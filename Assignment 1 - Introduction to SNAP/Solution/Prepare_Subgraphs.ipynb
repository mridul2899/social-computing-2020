```python
fb = open('facebook_combined.txt', 'r')
fb_edges = fb.readlines()
fb_nodes = set()
```

```python
for edge in fb_edges:
    node1, node2 = edge.strip().split()
    node1 = int(node1)
    node2 = int(node2)
    if node1 % 5 != 0:
        fb_nodes.add(node1)
    if node2 % 5 != 0:
        fb_nodes.add(node2)
```

```python
G1 = snap.TUNGraph.New()
for node in fb_nodes:
    G1.AddNode(node)
```

```python
for edge in fb_edges:
    node1, node2 = edge.strip().split()
    node1 = int(node1)
    node2 = int(node2)
    if node1 % 5 == 0 or node2 % 5 == 0:
        continue
    G1.AddEdge(node1, node2)
```

```python
snap.SaveEdgeList(G1, "./subgraphs/facebook.elist")
```

```python
fb.close()
```

```python
file1 = open('./subgraphs/amazon.elist', 'r')
file2 = open('./subgraphs/amazon_m.elist', 'r')
nodes1 = set()
nodes2 = set()
lines1 = file1.readlines()
lines2 = file2.readlines()
for line in lines1[3:]:
    node1, node2 = [int(i) for i in line.strip().split()]
    nodes1.add(node1)
    nodes1.add(node2)
for line in lines2:
    node1, node2 = [int(i) for i in line.strip().split()]
    nodes2.add(node1)
    nodes2.add(node2)
nodes1 = list(nodes1)
nodes2 = list(nodes2)
nodes1.sort()
nodes2.sort()
for i in range(len(nodes1)):
    if (nodes1[i] - nodes2[i] != 0):
        print(nodes1[i] - nodes2[i])
```

```python
edges1 = []
edges2 = []
for line in lines1[3:]:
    node1, node2 = [int(i) for i in line.strip().split()]
    if (node1 > node2):
        node1, node2 = node2, node1
    edges1.append((node1, node2))
for line in lines2:
    node1, node2 = [int(i) for i in line.strip().split()]
    if (node1 > node2):
        node1, node2 = node2, node1
    edges2.append((node1, node2))
```

```python
def Sort_Tuple(tup):  
  
    # reverse = None (Sorts in Ascending order)  
    # key is set to sort using second element of  
    # sublist lambda has been used  
    tup.sort(key = lambda x: x[0])  
    return tup
```

```python
edges1 = Sort_Tuple(edges1)
```

```python
edges2 = Sort_Tuple(edges2)
```
